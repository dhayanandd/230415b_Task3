```python
# Cleaning
```

```python
from sklearn.impute import SimpleImputer

df_cleaned = df.copy()

# Identify numeric and categorical columns
num_cols = df_cleaned.select_dtypes(include=['int64','float64']).columns
cat_cols = df_cleaned.select_dtypes(include='object').columns

# --- Handle missing numeric columns ---
num_imputer = SimpleImputer(strategy='median')
df_cleaned[num_cols] = num_imputer.fit_transform(df_cleaned[num_cols])

# --- Handle missing categorical columns ---
cat_imputer = SimpleImputer(strategy='most_frequent')
df_cleaned[cat_cols] = cat_imputer.fit_transform(df_cleaned[cat_cols])
```

```python
print("Missing values left:", df_cleaned.isnull().sum().sum())
print("Shape of cleaned dataset:", df_cleaned.shape)

df_cleaned.head()
```

```python
# Function for IQR capping
def iqr_cap(s, k=2.0):
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - k*iqr
    upper = q3 + k*iqr
    return s.clip(lower, upper)

# Apply to Price, Landsize, BuildingArea
for col in ['Price','Landsize','BuildingArea']:
    if col in df_cleaned.columns:
        df_cleaned[col] = iqr_cap(df_cleaned[col], k=2.0)
```

```python
# Make a copy
df_cleaned = df.copy()

# Strip whitespace from strings
for col in df_cleaned.select_dtypes(include='object'):
    df_cleaned[col] = df_cleaned[col].astype(str).str.strip()

# Parse Date
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], dayfirst=True, errors='coerce')

# Postcode as string
if 'Postcode' in df_cleaned.columns:
    df_cleaned['Postcode'] = df_cleaned['Postcode'].astype(str)
```

```python
# Sale year/month
df_cleaned['SaleYear'] = df_cleaned['Date'].dt.year.fillna(df_cleaned['Date'].dt.year.median())
df_cleaned['SaleMonth'] = df_cleaned['Date'].dt.month.fillna(df_cleaned['Date'].dt.month.median()).astype(int)

# Property age
df_cleaned['PropertyAge'] = (df_cleaned['SaleYear'] - df_cleaned['YearBuilt']).clip(lower=0)

# Price per sqm
df_cleaned['Price_per_sqm'] = df_cleaned['Price'] / (df_cleaned['BuildingArea'].replace(0, np.nan))
df_cleaned['Price_per_sqm'] = df_cleaned['Price_per_sqm'].clip(
    df_cleaned['Price_per_sqm'].quantile(0.01),
    df_cleaned['Price_per_sqm'].quantile(0.99)
)
```

```python
print("Nulls remaining:", int(df_cleaned.isnull().sum().sum()))
print("Shape:", df_cleaned.shape)

df_cleaned.to_csv("melbourne_cleaned.csv", index=False)
print("Saved cleaned dataset as melbourne_cleaned.csv")
```
